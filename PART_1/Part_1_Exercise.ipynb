```python
# 1
def prime_checker():
    num = int(input("Enter a number: "))
    if num > 1 and all(num % i != 0 for i in range(2, int(num**0.5) + 1)):
        print("This is a prime number.")
    else:
        print("This is not a prime number.")
prime_checker()
```

```python
# 2
def fibonacci_sequence():
    n = int(input("How many Fibonacci numbers to calculate? "))
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b
    print()
print("\n Fibonacci Sequence")
fibonacci_sequence()
```

```python
# 4
def reverse_string():
    user_input = input("Enter a string: ")
    print("Reversed string:", user_input[::-1])
print("\n Reverse String")
reverse_string()
```

```python
# 5
def min_max_values(iterable):
    return min(iterable), max(iterable)
def print_min_max():
    numbers = [40, 6, 0, 3, 4, 30, 22, 1, -1]
    min_val, max_val = min_max_values(numbers)
    print("Min value is:", min_val)
    print("Max value is:", max_val)
print("\n Min and Max Values")
print_min_max()
```

```python
# 6
import random
def guess_the_number():
    secret_number = random.randint(1, 10)
    while True:
        guess = int(input("Guess the number (1-10): "))
        if guess == secret_number:
            print("Congratulations! You make a good choice >.<.")
            break
        elif guess < secret_number:
            print("Too low! Try again.")
        else:
            print("Too high! Try again.")
print("\n Guess the Number ")
guess_the_number()
```

```python
# 7
def password_checker():
    correct_password = "MaxXx0"
    while True:
        password = input("Enter the password: ")
        if password == correct_password:
            print("Password Accept!")
            break
        else:
            print("Incorrect password. Try again.")
print("\nPassword Checker")
password_checker()
```
