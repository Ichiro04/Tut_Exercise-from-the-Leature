```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
number = int(input("Enter a number"))
if is_prime(number):
    print("This is a prime number")
else:
    print("This is not a prime number")
```

```python
def generate_fibonacci(n):
    fibonacci_sequence = [0, 1]
    for i in range(2, n):
        next_number = fibonacci_sequence[-1] + fibonacci_sequence[-2]
        fibonacci_sequence.append(next_number)
    return fibonacci_sequence[:n]  # Return only the first n numbers
num_terms = int(input("Enter how many numbers in the Fibonacci sequence to calculate: "))
if num_terms <= 0:
    print("Please enter a positive integer.")
else:
    fibonacci_sequence = generate_fibonacci(num_terms)
    print("Fibonacci sequence:")
    print(fibonacci_sequence)
```

```python
def reverse_string():
    user_defined = input("Enter a string: ")
    reversed_string = user_defined[::-1]
    return reversed_string
reversed_result = reverse_string()
print("Reversed string:", reversed_result)
```

```python
import numpy as np

# Temperature data in Celsius
temperature_data = [22, 24, 19, 21, 23, 25, 26, 27, 24, 23, 20, 22, 21, 19, 18, 22, 23, 24, 26, 28, 30, 27, 25, 26, 24, 22, 21, 20, 19, 23]

# 1) Create a NumPy array
temperature_array = np.array(temperature_data)

# 2) Find the minimum, maximum, and average temperature
min_temp = np.min(temperature_array)
max_temp = np.max(temperature_array)
avg_temp = np.mean(temperature_array)

print("Minimum temperature:", min_temp)
print("Maximum temperature:", max_temp)
print("Average temperature:", avg_temp)

# 3) Convert the temperatures from Celsius to Fahrenheit
fahrenheit_array = (temperature_array * 9/5) + 32
print("Temperatures in Fahrenheit:", fahrenheit_array)

# 4) Find all days where the temperature was above 25C
days_above_25 = temperature_array[temperature_array > 25]
print("Days with temperature above 25C:", days_above_25)

# 5) Count how many days had temperatures below 20C
days_below_20 = np.sum(temperature_array < 20)
print("Number of days with temperature below 20C:", days_below_20)

# 6) Sort the temperature data in ascending order
sorted_temperatures = np.sort(temperature_array)
print("Sorted temperatures:", sorted_temperatures)

# 7) Reshape the 1D array into a 5x6 2D array representing weeks
reshaped_array = temperature_array.reshape(5, 6)
print("Reshaped 5x6 array (weeks and days):")
print(reshaped_array)
```
